```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data = pd.read_csv('retail_sales_dataset.csv')
data.head(10)
```

```
data['Date'] = pd.to_datetime(data['Date'])
data['Month'] = data['Date'].dt.month
data.head()
```

```
women_df = data[data['Gender'] == 'Female']
men_df = data[data['Gender'] == 'Male']

men_amount_mean = men_df['Total Amount'].mean()
men_amount_median = men_df['Total Amount'].median()
women_amount_mean = women_df['Total Amount'].mean()
women_amount_median = women_df['Total Amount'].median()

print(f'Men Total Amount Mean: {men_amount_mean}\t Median: {men_amount_median}')
print(f'Men Total Amount Median: {men_amount_median}\t Median: {women_amount_median}')
print(f'Difference betmeen Mean: {abs(men_amount_mean - women_amount_mean)}\t Median:{abs(men_amount_mean - women_amount_mean)}')
```

```
age_amount_per_year_men = pd.DataFrame(men_df.groupby(['Age'])['Total Amount'].mean()).reset_index()
age_amount_per_year_women = pd.DataFrame(women_df.groupby(['Age'])['Total Amount'].mean()).reset_index()
age_amount_per_year_diff = pd.DataFrame({'Age': age_amount_per_year_men['Age'],"Total Amount": abs(age_amount_per_year_men['Total Amount'] - age_amount_per_year_women['Total Amount'])})

figure, ax = plt.subplots(3, 1, figsize=(15, 10))
sns.barplot(data = age_amount_per_year_men, x = 'Age', y = 'Total Amount', ax=ax[0])
ax[0].set_title('Men age mean Amount per Year')
ax[0].grid()
sns.barplot(data = age_amount_per_year_women, x = 'Age', y = 'Total Amount', ax=ax[1])
ax[1].set_title('Women age mean Amount per Year')
ax[1].grid()
sns.barplot(data = age_amount_per_year_diff, x = 'Age', y = 'Total Amount', ax=ax[2])
ax[2].set_title('Difference between Men and Women age mean Amount per Year')
ax[2].grid()
plt.show()
```
