```
import matplotlib.pyplot as plt
import numpy as np

data_1981_1990 = [470, 300, 250, 250, 190, 140, 40, 60, 120, 220, 480, 400]
data_1991_2000 = [450, 420, 420, 340, 140, 120, 50, 70, 60, 400, 490, 410]
data_2001_2010 = [420, 410, 370, 280, 120, 110, 60, 60, 60, 350, 630, 420]

bulan = np.arange(1, 13)

lebar_bar = 0.25

posisi1 = bulan - lebar_bar
posisi2 = bulan
posisi3 = bulan + lebar_bar

fig, ax = plt.subplots(figsize=(12, 7))

rects1 = ax.bar(posisi1, data_1981_1990, lebar_bar, label='1981 - 1990', color='#1f77b4') # Biru
rects2 = ax.bar(posisi2, data_1991_2000, lebar_bar, label='1991 - 2000', color='#d62728') # Merah
rects3 = ax.bar(posisi3, data_2001_2010, lebar_bar, label='2001 - 2010', color='#2ca02c') # Hijau

ax.set_xlabel('Bulan', fontweight='bold')
ax.set_ylabel('Curah Hujan (mm)', fontweight='bold')
ax.set_title('Visualisasi Data Curah Hujan per Periode', fontsize=16)

ax.set_xticks(bulan)
ax.set_xticklabels(bulan)
ax.set_ylim(0, 750)

ax.yaxis.grid(True, linestyle='--', which='major', color='grey', alpha=.25)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=3, fancybox=True, shadow=True)

fig.tight_layout(rect=[0, 0, 1, 0.96])

plt.show()
```

```
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

waktu_labels = [
    "00.00 UTC", "03.00 UTC", "06.00 UTC", "09.00 UTC",
    "12.00 UTC", "15.00 UTC", "18.00 UTC", "21.00 UTC", "00.00 UTC"
]
x_axis = range(len(waktu_labels))

data_asimilasi = [23.5, 25.5, 25.6, 25.3, 22.5, 22.3, 23.0, 22.0, 25.8]
data_observasi = [23.1, 25.2, 24.0, 24.0, 24.2, 24.3, 24.0, 23.0, 23.2]
data_non_asimilasi = [23.8, 25.6, 25.3, 25.5, 24.9, 25.1, 24.6, 24.3, 25.9]

fig, ax = plt.subplots(figsize=(11, 6))

ax.plot(x_axis, data_asimilasi, label='Asimilasi', marker='o', color='#0066cc', linewidth=2)
ax.plot(x_axis, data_observasi, label='Observasi', marker='o', color='#f0c800', linewidth=2)
ax.plot(x_axis, data_non_asimilasi, label='Non Asimilasi', marker='o', color='#cc0000', linewidth=2)

ax.set_title('Suhu Permukaan', fontsize=14, pad=10)

ax.set_ylabel('Celcius')
ax.set_ylim(20, 27.5)
ax.set_yticks(np.arange(21, 28, 1)) # Ticks dari 21 sampai 27

ax.set_xlabel('Waktu', labelpad=10)
ax.set_xticks(x_axis)
ax.set_xticklabels(waktu_labels)
ax.tick_params(axis='x', which='major', pad=5) # Memberi jarak antara label dan garis

plt.text(0.3, -0.12, '24 Juli 2013', ha='center', va='center', transform=ax.transAxes)
plt.text(0.9, -0.12, '25 Juli 2013', ha='center', va='center', transform=ax.transAxes)

ax.yaxis.grid(True, linestyle='-', which='major', color='lightgray', alpha=0.7)
ax.set_axisbelow(True)

ax.legend(loc='center right', bbox_to_anchor=(1, 0.4), frameon=True)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout(rect=[0, 0.05, 1, 1])

plt.show()
```
